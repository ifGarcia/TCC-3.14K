```python
dataTeste = dataframeDormindo.append(dataframeNormal, ignore_index=True) #data frame base recebe os dados
```

```python
dataTeste = dataTeste.reindex(np.random.permutation(dataTeste.index))
dataTeste = dataTeste.reset_index(drop=True)
```

```python
dataTeste
```

## Treinar essa budega

```python
from sklearn.model_selection import train_test_split
```

```python
#Separa os dados para trinar os modelos
X = dataTeste.drop('tag',axis=1)
y = dataTeste['tag']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
#Pega os modelos para serem treinados
svc_model = SVC()
lr_model = LogisticRegression()
```

```python
#Treina os modelos
svc_model.fit(X_train,y_train)
lr_model.fit(X_train,y_train)
```

## Avaliar o modelo

** Agora obtenha previsões do modelo e crie uma matriz de confusão e um relatório de classificação. **

```python
#Envia os dados para o modelo de previsão
svc_predictions = svc_model.predict(X_test)
lr_predictions = lr_model.predict(X_test)
```

```python
X_test.count()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,svc_predictions))
print(confusion_matrix(y_test,lr_predictions))
```

```python
print(classification_report(y_test,svc_predictions))
print(classification_report(y_test,lr_predictions))
```
