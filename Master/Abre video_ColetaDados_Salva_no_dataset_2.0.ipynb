```python
import cv2
import dlib
import pandas as pd
import os
```

```python
dataframe_base = pd.read_csv('Base_Normal.cvs') # #carrega dataframe base pd.DataFrame()
data_error = pd.DataFrame()
path_Video = '../Videos/Normal/' #MovimentoNormal
#Configuração do video
seg_Video = 5
fps_Video = 30
frames_totais = seg_Video*fps_Video
colunas_Totais = frames_totais*68*2
```

```python
#Loop lendo os nomes dos arquivos
for nome_Arquivo in os.listdir(path_Video):
    #para cada video no diretorio coletar os dasdos e passar pro dataframe_Base
    print (nome_Arquivo)
    coletaDados(path_Video+nome_Arquivo)
```

```python
def rotaciona(img):
    (alt, lar) = img.shape[:2] #captura altura e largura
    centro = (lar // 2, alt // 2) #acha o centro
    M = cv2.getRotationMatrix2D(centro, 90, 1.0)
    return cv2.warpAffine(img, M, (lar, alt))

def coletaDados(arquivo_Video):
    print (arquivo_Video)
    
    #Abre o video e configura
    global dataframe_base
    global data_error
    video_capture = cv2.VideoCapture(arquivo_Video) #carrega o video na variavel
    video_capture.set(cv2.CAP_PROP_FPS, 30) #Seta o FPS para 30
    
    #Inicia variaveis e dependencias
    detector = dlib.get_frontal_face_detector() #Mode de predição de rosto
    predictor = dlib.shape_predictor("../Dependencias/shape_predictor_68_face_landmarks.dat") #Modelo de predição dos postos do rosto
    new_frames_totais = frames_totais #caso apenas um video, o max se altera para o proximo "corte"
    dataframe_Linha = pd.DataFrame() #criar data frame vazio
    video_frame_total = video_capture.get(cv2.CAP_PROP_FRAME_COUNT)
    semRosto = 0           #controle de frame sem rosto indentificado
    
    print (new_frames_totais)
    
    while video_capture.grab(): #enquanto tiver frames
        
        ret, frame = video_capture.retrieve() #pega proximo quadro
        video_frame_atual = video_capture.get(cv2.CAP_PROP_POS_FRAMES)
        print (video_frame_atual)
        frame = rotaciona(frame)
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) #passa a imagem para cinza
        #clahe_image = clahe.apply(gray)                #aplica uma configuração para melhor detecção
        #cv2.imshow("image", frame) #mostra a imagem

        detections = detector(gray, 1) #detecta as faces na imagem

        if detections: #se rosto detectado
            print ('Com Rosto')
            ultimo_frame = pd.DataFrame() #criar data frame vazio
            #for k,d in enumerate(detections): #For each detected face
            shape = predictor(gray, detections[0]) #Get coordinates
            for i in range(0,68): #There are 68 landmark points on each face
                ultimo_frame = pd.concat([ultimo_frame,pd.DataFrame({str(i+1)+'.x':[shape.part(i).x] ,str(i+1)+'.y':[shape.part(i).y]})],axis=1)
            dataframe_Linha = pd.concat([dataframe_Linha,ultimo_frame],axis=1)
        else:
            print ('Sem Rosto')
            dataframe_Linha = pd.concat([dataframe_Linha,ultimo_frame],axis=1)
            semRosto+=1
    
        if video_frame_atual == new_frames_totais: #se coletou o esperado
            print ('coletou o esperado')
            print (video_frame_atual)
            dataframe_Linha['tag'] = 'normal'
            dataframe_Linha.columns = dataframe_base.columns
            data_error = dataframe_Linha
            dataframe_base = dataframe_base.append(dataframe_Linha,ignore_index=True) #data frame base recebe os dados
            if video_frame_total-video_frame_atual < new_frames_totais: #se coletou todo o possivel
                print ('coletou tudo e ja era!')
                video_capture.release() #fecha o arquivo do video carregado
                break
            else:
                new_frames_totais += frames_totais #dobra limite para o proximo corte
                dataframe_Linha = pd.DataFrame() #zera o data frame linha
    
    
    
    
    video_capture.release() #fecha o arquivo do video carregado
```
