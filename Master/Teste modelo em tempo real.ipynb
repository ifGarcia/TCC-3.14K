```python
import cv2
import dlib
import pandas as pd
import os
import pickle
from xgboost import XGBClassifier
```

```python
xgbr_model_load = pickle.load(open('xgbr_model', 'rb'))
```

```python
xgbr_model_load.feature_names
```

```python
data_error = []
```

```python
video_capture = cv2.VideoCapture(0) #carrega o video na variavel
video_capture.set(cv2.CAP_PROP_FPS, 30) #Seta o FPS para 30

#Inicia variaveis e dependencias
detector = dlib.get_frontal_face_detector() #Mode de predio de rosto
predictor = dlib.shape_predictor("../Dependencias/shape_predictor_68_face_landmarks.dat") #Modelo de predio dos postos do rosto
dataframe_Linha = [] #criar data frame vazio
primeiroFrame = False

global data_error

while True:
    ret, frame = video_capture.read() #pega proximo quadro
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) #passa a imagem para cinza
    #cv2.imshow("image", frame) #mostra a imagem

    detections = detector(gray, 1) #detecta as faces na imagem

    if detections: #se rosto detectado
        cv2.putText(frame, "Pegando rosto", (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
        primeiroFrame = True
        ultimo_frame = [] #criar data frame vazio
        shape = predictor(gray, detections[0]) #Get coordinates
        for i in range(0,68): #There are 68 landmark points on each face
            ultimo_frame.extend([shape.part(i).x, shape.part(i).y])
        dataframe_Linha.extend(ultimo_frame)
    else:
        if primeiroFrame:
            dataframe_Linha.extend(ultimo_frame)
            
    cv2.putText(frame, str(len(dataframe_Linha)), (10, 130), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
    data_error = dataframe_Linha
    if len(dataframe_Linha) > 20400:
        dataframe_Linha = pd.DataFrame(dataframe_Linha[135:])  #apaga frame 1 - 136 colunas
        data_error = dataframe_Linha
        #cv2.putText(frame, str(modelo.predict(X_test)), (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
        cv2.putText(frame, str(xgbr_model_load.predict(dataframe_Linha)), (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
    else:
        #cv2.putText(frame, str(modelo.predict(X_test)), (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
        cv2.putText(frame, "Nada coletado ainda", (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
        
    if cv2.waitKey(1) & 0xFF == ord('q'): #Exit program when the user presses 'q'
        video_capture.release()
        cv2.destroyAllWindows()
        break
    
    cv2.imshow("image", frame) #mostra a imagem
    
video_capture.release() #fecha o arquivo do video carregado
```
